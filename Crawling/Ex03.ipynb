```python
## CLASS
url = 'https://www.hsd.co.kr/menu/menu_list'
driver = wb.Chrome()
driver.get(url)

soup= bs(driver.page_source, 'lxml')

menu=soup.select('div.item-text > h4')

price = soup.select('div.item-price > strong')

menu_list = []
price_list = []
rank_list = []

for i in range(len(menu)):
    menu_list.append(menu[i].text)
    price_list.append(price[i].text)
    rank_list.append(i+1)

int(price_list[0].replace(',', ''))

for i in range(len(price_list)):
    price_list[i] = int(price_list[i].replace(',', ''))

dic = {'메뉴이름' : menu_list, '가격' : price_list, '번호' : rank_list}

df = pd.DataFrame(dic)

df.set_index('번호', inplace =True)

df =df.sort_values('가격', ascending = False) #ascending 내림차순(False), 오름차순(True)
df.reset_index(drop=True, inplace = True)
df2 = pd.DataFrame(rank_list, columns=['번호'])
df = pd.concat([df, df2], axis=1).set_index('번호')
ㅠ
df
```
