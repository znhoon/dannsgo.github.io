# Question3

```python
import requests as req  # requests 라이브러리 불러오기
from bs4 import BeautifulSoup as bs # BeautifulSoup 라이브러리 불러오기
import pandas as pd
import time

driver = wb.Chrome() # 크롬 웹 페이지 연결
url = 'http://corners.gmarket.co.kr/Bestsellers'
driver.get(url) # 크롬에 url 연결
soup = bs(driver.page_source, 'lxml') # 파싱
```

```python
title_list = []
price_list = []
rank_list = []

img = driver.find_elements_by_css_selector('img.lazy')

for i in range(20):
    img = driver.find_elements_by_css_selector('img.lazy') 
    img[i].click() ## 클릭
    time.sleep(0.4) ## 웹페이지 바뀌는 시간
    
    soup = bs(driver.page_source, 'lxml') # 파싱
    title = soup.select_one('h1.itemtit') 
    price = soup.select_one('strong.price_real')
    
    title_list.append(title.text) 
    price_list.append(price.text)
    rank_list.append(i+1) 
    
    driver.back()
    time.sleep(0.4)
```

```python
dic = {'상품명' : title_list, '가격' : price_list, '랭킹' : rank_list} # 딕셔너리 생성
df = pd.DataFrame(dic) # 데이터프레임 생성
df.set_index('랭킹', inplace =True) # index 컬럼 => 번호 컬럼으로 덮어씌움
pd.set_option('display.max_rows', None) # 행 전부 출력, None : 최대
df
```

## 문제1

```python
h = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36'}
url = 'https://www.melon.com/chart/index.htm'
res = req.get(url, headers = h)
soup = bs(res.text, 'lxml')

## 노래제목
song = soup.select('div.ellipsis.rank01 > span > a')
## 가수
singer = soup.select('div.ellipsis.rank02 > a:first-child')

## 리스트
song_list = []
singer_list = []
cnt_list = []

## 범위 == TOP100
for i in range(len(song)):
    song_list.append(song[i].text)
    singer_list.append(singer[i].text)
    cnt_list.append(i+1)

## DataFrame
dic = {'노래 제목' : song_list, '가수' : singer_list, '순위' : cnt_list}
df1 = pd.DataFrame(dic)
df1.set_index('순위', inplace = True)
pd.set_option('display.max_rows', None) # 리스트 전체 출력
df1
```

```python
df1.to_csv('./melon_chart', encoding='utf-8') # 현재 위치에서 '파일명'으로 csv로 변환해 저장
```

```python
import requests as req  # requests 라이브러리 불러오기
from bs4 import BeautifulSoup as bs # BeautifulSoup 라이브러리 불러오기
import pandas as pd
import time

driver = wb.Chrome() # 크롬 웹 페이지 연결
url = 'http://www.youtube.com/results?search_query=%EC%9A%B0%EC%99%81%EA%B5%B3'
driver.get(url) # 크롬에 url 연결
soup = bs(driver.page_source, 'lxml') # 파싱

## 영상 제목
title = soup.select('#container > h1 > yt-formatted-string')
## 조회수
num = soup.select('#count > ytd-video-view-count-renderer > span.view-count.style-scope.ytd-video-view-count-renderer')

## 리스트
title_list = []
num_list = []
cnt_list = []

body = driver.find_element_by_css_selector('body')
## 범위 == 100 이상
for i in range(100):
    title_list.append(title[i].text)
    num_list.append(num[i].text)
    cnt_list.append(i+1)
    print('%s번째 리스트가 저장 중입니다.'%len(cnt_list))
    time.sleep(3)
    body.send_keys(Keys.END)
```

```python
## DataFrame
dic = {'영상 제목' : title_list, '조회수' : num_list, '번호' : cnt_list}
df = pd.DataFrame(dic)
df.set_index('번호', inplace = True)
pd.set_option('display.max_rows', None) # 리스트 전체 출력
df
```
