### matplotlib란?
matplotlib은 파이썬에서 데이터를 차트나 플롯(Plot)으로 그려주는 시각화 라이브러리 라인 플롯, 바 차트, 파이차트, 히스토그램, Bot Plot, Scatter Plot 등을 비롯한 다양한 차트와 플롯 스타일을 지원

#### 라이브러리 사용

```python
import matplotlib.pyplot as plt
```

```python
y=[2,4,6]

plt.plot(y)
plt.show()
```

```python
x=[1,2,3]
y=[2,4,6]

plt.plot(x, y)
plt.show()
```

### Line style

```python
y=[1,3,7]
plt.plot(y, ls = '-') # ls == line style의 약자
plt.show()
```

#### Line width
lw = 그래프 굵기(숫자로 지정)

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x, y, lw = 4)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x, y, lw = 4, ls = '--') # lw와 ls 동시에 적용 가능
plt.show()
```

### Marker 

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x, y, lw = 1, ls = '--', marker = '|') 
plt.show()
```

### Line color

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x, y, lw = 1, ls = '--', marker = '|', c = 'c') 
plt.show()
```

### Marker 스타일

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x, y, lw = 1, ls = '--', marker = 'o', c = 'c', ms = 10) 
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x, y, lw = 1, ls = '--', marker = 'o', c = 'c', ms = 10, mfc = 'b') 
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x, y, lw = 1, ls = '--', marker = 'o', c = 'c', ms = 10, mfc = 'b', mec = 'b') 
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x, y, lw = 1, ls = '--', marker = 'o', c = 'c', ms = 10, mfc = 'b', mec = 'y', mew = 2) 
plt.show()
```

#### 틱 설정
- 플롯이나 차트에서 축상의 위치 표시 지점을 틱(tick)이라고 함
- 틱에 써진 숫자 혹은 글자 틱 라벨(tick label)
- 틱의 위치나 틱 라벨은 Matplotlib가 자동으로 지정, 수동으로 설정하고 싶다면 xticks, yticks 명령을 이용

```python
x = np.arange(7) # 6까지
y = [1, 4, 5, 8, 9, 5, 3]

plt.plot(x, y, ls = ':', marker = 'D', c = 'r', mfc = 'b') 
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.show()
```

### 그리드

```python
x = np.arange(7) # 6까지
y = [1, 4, 5, 8, 9, 5, 3]

plt.plot(x, y, ls = ':', marker = 'D', c = 'r', mfc = 'b') 
plt.grid()
plt.show()
```

### 여러 개의 플롯

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
z = [3, 6, 9, 12]

plt.plot(x)
plt.plot(y)
plt.plot(z)
plt.grid()
plt.show()
```

### 범례 표시

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
z = [3, 6, 9, 12]

plt.plot(x, label = 'X')
plt.plot(y, label = 'Y')
plt.plot(z, label = 'Z')
plt.legend() # 범례를 적용
plt.grid()
plt.show()
```
